```python
import yfinance as yf
```

```python
# Get S&P 500 historical data
sp500 = yf.Ticker("^GSPC")
sp500 = sp500.history(period="max")
```

```python
# Delete unnecessary columns
del sp500["Dividends"]
del sp500["Stock Splits"]
```

```python
sp500
```

```python
# Create 'Tomorrow' column by shifting 'Close' prices by -1
sp500["Tomorrow"] = sp500["Close"].shift(-1)
```

```python
sp500
```

```python
# Create 'Target' column: 1 if tomorrow's close > today's close, else 0
sp500["Target"] = (sp500["Tomorrow"] > sp500["Close"]).astype(int)
```

```python
sp500
```

```python
# Remove all data before 1990
sp500 = sp500.loc["1990-01-01":].copy()
sp500
```

```python
# Using RandomForestClassifier from sklearn because of its robustness and effectiveness for classification tasks
# What randomforestclassifier from sklearn is: A machine learning algorithm that builds multiple decision trees with randomized parameters 
# and merges them together to get a more accurate and stable prediction.
# Reasons:
# 1. More resistant to overfitting 
# 2. Runs relatively quickly on large datasets
# 3. Handles non-linear data well, perfect for unpredictable stock market trends
from sklearn.ensemble import RandomForestClassifier

# Initialize the RandomForestClassifier model
# n_estimators: Number of decision trees in the forest that we want to train (might want to tune this later, maybe try higher values for better accuracy)
# min_samples_split: Minimum number of samples required to split an internal node (helps prevent overfitting by ensuring nodes have enough data)
# The higher we set min_samples_split, the less accurate it will be, but the less it will overfit
# random_state: Ensures reproducibility of results by setting a seed for random number generation, 1 makes sure we get same results every time we run the code
model = RandomForestClassifier(n_estimators=100, min_samples_split=30, random_state=1)

# Split up data into train and test sets
train = sp500.iloc[:-100] # All but last 100 days for training
test = sp500.iloc[-100:] # Last 100 days for testing

predictors = ["Close", "Volume", "Open", "High", "Low"] # Features to use for prediction
model.fit(train[predictors], train["Target"]) # Train the model, using the predictors to predict the Target
```

```python
from sklearn.metrics import precision_score # To test model's accuracy

preds = model.predict(test[predictors]) # Make predictions on the test set
preds
```

```python
import pandas as pd
# Convert predictions to a pandas Series with the same index as the test set
preds = pd.Series(preds, index=test.index) 
```

```python
preds
```

```python
precision_score(test["Target"], preds) # Calculate precision of the predictions
```

```python
combined = pd.concat([test["Target"], preds], axis=1)
combined.plot()
```

```python
def predict(train, test, predictors, model):
    model.fit(train[predictors], train["Target"])
    preds = model.predict(test[predictors])
    preds = pd.Series(preds, index=test.index, name="Predictions")
    combined = pd.concat([test["Target"], preds], axis=1)
    return combined
```

```python
# Backtesting function to evaluate model performance over multiple time periods
# data: our sp500 dataframe
# model: our RandomForestClassifier model
# predictors: list of feature column names to use for prediction
# start: index to start backtesting from (default 2500)
    # Every trading year is about 250 days, so starting at 2500 means we train our first model with 10 years of data
# step: number of days to increment for each backtest iteration (default 250)
    # This means we are going to take 10 years of data and predict values for the 11th year, then we take 11 years of data and predict values for the 12th year, and so on
def backtest(data, model, predictors, start=2500, step=250):
    all_predictions = [] # List of dataframes from each year

    for i in range(start, data.shape[0], step):
        train = data.iloc[0:i].copy()
        test = data.iloc[i:(i+step)].copy()

        predictions = predict(train, test, predictors, model)
        all_predictions.append(predictions)
    return pd.concat(all_predictions)
```

```python
predictions = backtest(sp500, model, predictors)
```

```python
predictions["Predictions"].value_counts()
```

```python
precision_score(predictions["Target"], predictions["Predictions"])
```

```python
# Comparing close price to last two, five, sixty, 250, and 1000 days
# Basically, adding features to compare the price to recent trends
horizons = [2, 5, 60, 250, 1000]
new_predictors = []

for horizon in horizons:
    rolling_averages = sp500.rolling(horizon).mean()

    ratio_column = f"Close_Ratio_{horizon}"
    sp500[ratio_column] = sp500["Close"] / rolling_averages["Close"]

    trend_column = f"Trend_{horizon}"
    sp500[trend_column] = sp500.shift(1).rolling(horizon).sum()["Target"]

    new_predictors += [ratio_column, trend_column]
```

```python
sp500 = sp500.dropna()
sp500
```

```python
model = RandomForestClassifier(n_estimators=200, min_samples_split=50, random_state=1)
```

```python
def predict(train, test, predictors, model):
    model.fit(train[predictors], train["Target"])
    preds = model.predict_proba(test[predictors])[:, 1]
    # Setting a threshold of 0.6 for classification
    # Reduces the number of days that the model predicts an increase, potentially improving precision
    preds[preds >= .6] = 1
    preds[preds < .6] = 0
    preds = pd.Series(preds, index=test.index, name="Predictions")
    combined = pd.concat([test["Target"], preds], axis=1)
    return combined
```

```python
predictions = backtest(sp500, model, new_predictors)
```

```python
predictions["Predictions"].value_counts()
```

```python
precision_score(predictions["Target"], predictions["Predictions"])
```
